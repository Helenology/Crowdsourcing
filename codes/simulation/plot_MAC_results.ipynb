```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data1 = pd.read_csv("./results/[K=1]MAC_data.csv")
data2 = pd.read_csv("./results/[K=2]MAC_data.csv")
data = pd.concat([data1, data2], axis=0)
data.shape
```

```python
data = data.sort_values(by=['seed', 'n', 'alpha'])
data.head()
```

```python
MAC = data.loc[:, ["n", "alpha", "M", "K", "seed", "OMR", 
                    "CMR(0.1)", "CMR(0.05)", "CMR(0.01)", 
                    "ELC(0.1)", "ELC(0.05)", "ELC(0.01)"]]
MAC_tab = MAC.groupby(['K', 'n', 'alpha']).mean()
```

```python
MAC_tab["OMR"] = MAC_tab["OMR"] * 100
MAC_tab["CMR(0.1)"] = MAC_tab["CMR(0.1)"] * 100
MAC_tab["ELC(0.1)"] = MAC_tab["ELC(0.1)"] * 100
MAC_tab["CMR(0.05)"] = MAC_tab["CMR(0.05)"] * 100
MAC_tab["ELC(0.05)"] = MAC_tab["ELC(0.05)"] * 100
MAC_tab["CMR(0.01)"] = MAC_tab["CMR(0.01)"] * 100
MAC_tab["ELC(0.01)"] = MAC_tab["ELC(0.01)"] * 100
MAC_tab
```

```python
MAC_tab = pd.DataFrame(MAC_tab)
MAC_tab = MAC_tab.reset_index()
MAC_tab
```

```python
MAC_tab.to_csv("./results/MAC_tab.csv")
```
