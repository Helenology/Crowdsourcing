```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data1 = pd.read_csv("./results/[K=1]RMSE_data.csv")
data2 = pd.read_csv("./results/[K=2]RMSE_data.csv")
data = pd.concat([data1, data2], axis=0)
data.shape
```

```python
data = data.sort_values(by=['seed', 'n', 'alpha'])
data.head()
```

```python
data.columns
```

```python
RMSE = data.loc[:, ['n', 'alpha', 'M', 'K', 
                    'init_beta',  'os_beta',  'ts_beta',  'ms_beta',  'oracle_beta', 
                    'init_sigma', 'os_sigma', 'oracle_sigma']]
RMSE_tab = RMSE.groupby(['K', 'n', 'alpha']).mean()
```

```python
RMSE_tab["init_beta"] = RMSE_tab["init_beta"] * 1000
RMSE_tab["os_beta"] = RMSE_tab["os_beta"] * 1000
RMSE_tab["ts_beta"] = RMSE_tab["ts_beta"] * 1000
RMSE_tab["ms_beta"] = RMSE_tab["ms_beta"] * 1000
RMSE_tab["oracle_beta"] = RMSE_tab["oracle_beta"] * 1000


RMSE_tab["init_sigma"] = RMSE_tab["init_sigma"] * 100
RMSE_tab["os_sigma"] = RMSE_tab["os_sigma"] * 100
RMSE_tab["oracle_sigma"] = RMSE_tab["oracle_sigma"] * 100

RMSE_tab
```

```python
RMSE_tab = pd.DataFrame(RMSE_tab)
RMSE_tab = RMSE_tab.reset_index()
RMSE_tab
```

```python
RMSE_tab.to_csv("./results/RMSE_tab.csv")
```
